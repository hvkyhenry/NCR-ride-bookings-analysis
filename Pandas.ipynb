**Adding/Removing column**

```python
coffee['price'] = 677
```

```python
import numpy as np
coffee['new_price'] = np.where(coffee['Coffee Type'] == 'Latte',5.66,4.33)
```

```python
coffee
```

```python
coffee.drop(columns = 'price',inplace=True)
# If you don't inplace=True, if you run coffee again you'll still see the column
```

```python
 print(coffee.head())
```

```python
coffee['revenue'] = coffee['Units Sold'] * coffee['new_price']
```

```python
print(coffee.head())
```

```python
coffee.rename(columns = {'new_price':'Price'})
```
